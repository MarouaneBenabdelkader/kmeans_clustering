```python
import pandas as pd
import numpy as np
import sklearn.metrics as sm
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
iris = datasets.load_iris()
# print the iris data
# print(iris)
# print(iris.data)
print(iris.target)
print(iris.feature_names)
print(iris.target_names)
```

```python
data = pd.DataFrame(iris.data)
print(data.head())
```

```python
data.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']
print(data.head())
```

```python
y =  pd.DataFrame(iris.target)
y.columns = ['Targets']
print(y.head())
```

```python
tab = []
for i in range(2,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    tab.append(kmeans.inertia_)
plt.plot(range(2,10), tab)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('tab')
plt.show()
```

```python
model = KMeans(n_clusters=3)
model.fit(data)
print(model.labels_)
```

```python
plt.scatter(data.Petal_Length, data.Petal_Width)
```

```python
colormap = np.array(['red', 'lime', 'black'])
plt.scatter(data.Petal_Length, data.Petal_Width, c=colormap[model.labels_], s=40)
```

```python
plt.scatter(data.Petal_Length, data.Petal_Width, c=colormap[y.Targets], s=40)
```

```python
model = pd.DataFrame(model.labels_)
model.columns = ["Targets"]
data = pd.concat([data, model], axis=1)
print(data.head())
```
